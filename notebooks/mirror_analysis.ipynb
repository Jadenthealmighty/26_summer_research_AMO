```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("/Users/jadenal-aidroos/Desktop/ROP_code/data/csv_raw/new_mirror_meas.csv")

```

```python
inc_angle = df["Incedence angle"].values
diag = df["Diag"].values
V = df["V"].values
H = df["H"].values
max_angle = df["Max Angle"].values
max_val = df["Max"].values
min_angle = df["Min Angle"].values
min_val = df["Min"].values
```

```python
# Baseline values
H_base = 1.369
V_base = 1.375
Max_base = 2.705
```

```python
plt.figure(figsize=(10,7))
plt.plot(inc_angle, max_val, color='red', markersize=10, label='Max')
plt.plot(inc_angle, min_val, color='blue', markersize=10, label='Min')
plt.axhline(y=Max_base, color='black', linestyle='--', label='Baseline Max')

plt.legend()
plt.xlabel('Incidence angle (in deg)')
plt.ylabel('Power (from multimeter in V)')
plt.title('Mirror measurements')

# V and H polarization:

plt.figure(figsize=(10,7))
plt.plot(inc_angle, H, color='red', markersize=10, label='P')
plt.plot(inc_angle, V, color='blue', markersize=10, label='S')
plt.axhline(y=H_base, color='red', linestyle='--', label='Baseline H')
plt.axhline(y=V_base, color='blue', linestyle='--', label='Baseline V')


plt.legend()
plt.xlabel('Incidence angle (in deg)')
plt.ylabel('Power (from multimeter in V)')
plt.title('Mirror measurements')

plt.show()
```

```python
VH_sum = V + H

tot = max_val + min_val


plt.figure(figsize=(10,7))
plt.plot(inc_angle, VH_sum, label='V+H')
plt.plot(inc_angle, diag, label='Diagonal')
plt.plot(inc_angle, max_val, label='Max')
plt.plot(inc_angle, tot, label='Min')



plt.legend()
plt.xlabel('Incidence angle (in deg)')
plt.ylabel('Power (from multimeter in V)')
plt.title('Mirror measurements')

plt.show()
```
